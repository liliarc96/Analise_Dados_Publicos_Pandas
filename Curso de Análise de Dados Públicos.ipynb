```python
#acessando o site
url = "https://sismpconsultapublica.mpsp.mp.br/ConsultarDistribuicao/ObterProcedimentosPorMembro"
p = requests.get(url)
s = bs(p.text, "html.parser")
#procura a tag option
lista = s.findAll("option")
nomes = lista[1:-5]
for name in nomes:
#nome e valor    
    print(name.string, name["value"])
```

```python
#número de nomes na string nomes
print(len(nomes))
```

```python
u = "http://filiaweb.tse.jus.br/filiaweb/portal/relacoesFiliados.xhtml"
p = requests.get(u)
s = bs(p.text, "html.parser")
itens = s.findAll("option")
partidos = [p["value"] for p in itens]
#tira os estados
partidos = partidos[:-27]
print(partidos)
```

```python
url = "http://agencia.tse.jus.br/estatistica/sead/eleitorado/filiados/uf/filiados_"
from urllib.request import urlretrieve
for p in partidos:
    u = f"{url}{p}_pb.zip"
    file = f"{p}_pb.zip"
    print(file)
    urlretrieve(u, file)
```

```python
#biblioteca de zip
import zipfile
import os
#no diretório atual
files = os.listdir(".")
#todos os que terminam com .zip
for f in files:
    if f.endswith(".zip"):
        print(f)
        #ler
        zip_ref = zipfile.ZipFile(f, "r")
        #extrair
        zip_ref.extractall("Filiados")
        #fechar
        zip_ref.close()
```

```python
from pytrends.request import TrendReq
pytrend = TrendReq(hl='BR')
candidatos = ['Bolsonaro', 'Marina Silva',
              'Alckmin', 'Ciro Gomes', 'Boulos']
pytrend.build_payload(kw_list=candidatos)
trends = pytrend.related_queries()
for p in trends:
  print (f'Rising {p}')
  print (trends[p]['rising'])
  print (f'Top {p}')
  print (trends[p]['top'])
  print ()
```
